```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv")
```

```python
df.head()
```

```python
#Y variable -> Spending Score(1-100)
#X variables -> Gender, Age, Annual Income($)
```

```python
#split train set and test set
from sklearn.model_selection import train_test_split

#defining X and Y variables
X = df.loc[:,df.columns != 'Spending Score (1-100)']
y =df['Spending Score (1-100)']

#partitioning the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train
```

```python
#import lazypredict
import lazypredict
```

```python
#importing regression class from lazypredict
from lazypredict.Supervised import LazyRegressor
```

```python
#defining our model using LazyRegressor
multiple_ML_model = LazyRegressor(verbose=0,ignore_warnings=True,predictions=True)
```

```python
#fitting the models along with predictiong with outputs o feach model
models,predictions = multiple_ML_model.fit(X_train,X_test,y_train,y_test)
```
